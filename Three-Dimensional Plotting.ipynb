## Three-dimensional Plotting

```python
import numpy as np
import matplotlib.pyplot as plt

### 3-Dimensional Scattered Graph
```

### 3-Dimensional Line Graph

```python
from mpl_toolkits import mplot3d

fig = plt.figure()

ax = plt.axes(projection ='3d')

z = np.linspace(0, 1, 100)
x = z * np.sin(25 * z)
y = z * np.cos(25 * z)

ax.plot3D(x, y, z, 'green')
ax.set_title('3D line plot geeks for geeks')
plt.show()
```

### 3-Dimensional Scattered Graph

```python
ax = plt.axes(projection='3d')

z = np.linspace(0, 1, 100)
x = z * np.sin(25*z)
y = z * np.cos(25*z)
c = x + y

ax.scatter(x, y, z, c=c)
ax.set_title('3D Scatter Plot')
plt.show()
```

### Surface Graphs

```python
x = np.outer(np.linspace(-2, 2, 10), np.ones(10))
y = x.copy().T
z = np.cos(x**2 + y**3)

fig = plt.figure()

ax = plt.axes(projection='3d')

ax.plot_surface(x, y, z, cmap='viridis', edgecolor='g')
ax.set_title('Surface Plot')
plt.show()
```

### Wireframes graph

```python
def f(x, y):
    return np.sin(np.sqrt(x**2 + y**2))

x = np.linspace(-1,5,10)
y = np.linspace(-1,5,10)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_wireframe(X, Y, Z, color='g')
ax.set_title('Wireframe')
```

### Contour Graph

```python
def f(x, y):
    return np.sin(np.sqrt(x**2 + y**2))

x = np.linspace(-10,10,40)
y = np.linspace(-10,10,40)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)

fig = plt.figure(figsize=(10, 8))
ax = plt.axes(projection='3d')

ax.plot_surface(X, Y, Z, cmap='cool', alpha=0.8)

ax.set_title('3D Contour Plot of Function(x, y) = sin(sqrt(x^2 + y^2))', fontsize=14)
ax.set_xlabel('x', fontsize=12)
ax.set_ylabel('y', fontsize=12)
ax.set_zlabel('z', fontsize=12)
plt.show()
```

### Plotting Surface Triangulations

```python
from matplotlib.tri import Triangulation

def f(x, y):
    return np.sin(np.sqrt(x**2 + y**2))

x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)

tri = Triangulation(X.ravel(), Y.ravel())

fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(111, projection='3d')

ax.plot_trisurf(tri, Z.ravel(), cmap='cool', edgecolor='none', alpha=0.8)
ax.set_title('Surface Triangulation Plot of Function(x, y) = sin(sqrt(x^2 + y^2))', fontsize=14)

ax.set_xlabel('x', fontsize=12)
ax.set_ylabel('y', fontsize=12)
ax.set_zlabel('z', fontsize=12)
 
plt.show()
```

### Plotting Mbius Strip

```python
R = 2

u = np.linspace(0, 2*np.pi, 100)
v = np.linspace(-1, 1, 100)
u, v = np.meshgrid(u, v)
x = (R + v*np.cos(u/2))*np.cos(u)
y = (R + v*np.cos(u/2))*np.sin(u)
z = v * np.sin(u/2)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.plot_surface(x, y, z, alpha=0.5)

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Mbius Strip')

ax.set_xlim([-3, 3])
ax.set_ylim([-3, 3])
ax.set_zlim([-3, 3])

plt.show()
```

## 3D Scatter Plotting

```python
import random
random.seed(42)

z = np.random.randint(100, size =(50))
x = np.random.randint(80, size =(50))
y = np.random.randint(60, size =(50))

fig = plt.figure(figsize = (10, 7))
ax = plt.axes(projection ="3d")

ax.scatter3D(x, y, z, color='g')
plt.title('Simple 3D Scatter Plot')
plt.show()
```

```python
random.seed(42)

z = 4 * np.tan(np.random.randint(10, size =(500))) + np.random.randint(100, size =(500))
x = 4 * np.cos(z) + np.random.normal(size = 500)
y = 4 * np.sin(z) + 4 * np.random.normal(size = 500)

fig = plt.figure(figsize = (16, 9))
ax = plt.axes(projection ="3d")

ax.grid(visible=True, color='grey', linestyle='-.', linewidth=0.3, alpha=0.2)

my_cmap = plt.get_cmap('hsv')

sctt = ax.scatter3D(x, y, z, alpha=0.8, c=(x+y+z), cmap=my_cmap, marker='^')

plt.title("simple 3D scatter plot")
ax.set_xlabel('X-axis', fontweight ='bold') 
ax.set_ylabel('Y-axis', fontweight ='bold') 
ax.set_zlabel('Z-axis', fontweight ='bold')
fig.colorbar(sctt, ax = ax, shrink = 0.5, aspect = 5)

plt.show()
```

## 3D Surface Plotting

```python
x = np.outer(np.linspace(-3, 3, 32), np.ones(32))
y = x.copy().T
z = (np.sin(x **2) + np.cos(y **2) )

fig = plt.figure(figsize=(14,9))
ax = plt.axes(projection='3d')

ax.plot_surface(x,y,z)
plt.show()
```

### Gradient surface Plot

```python
x = np.outer(np.linspace(-3, 3, 32), np.ones(32))
y = x.copy().T
z = (np.sin(x **2) + np.cos(y **2) )

fig = plt.figure(figsize=(14,9))
ax = plt.axes(projection='3d')

my_cmap = plt.get_cmap('hot')

surf = ax.plot_surface(x,y,z, cmap=my_cmap, edgecolor='none')

fig.colorbar(surf, ax=ax, shrink=0.5, aspect=5)

ax.set_title('Surface Plor')

plt.show()
```

### 3D surface Plot having 2D contour plot projections

```python
x = np.outer(np.linspace(-3, 3, 32), np.ones(32))
y = x.copy().T
z = (np.sin(x **2) + np.cos(y **2) )

fig = plt.figure(figsize=(14,9))
ax = plt.axes(projection='3d')

my_cmap = plt.get_cmap('hot')

surf = ax.plot_surface(x,y,z, rstride=8, cstride=8, alpha=0.8, cmap=my_cmap)

cset = ax.contourf(x,y,z, zdir='z', offset=np.min(z), cmap=my_cmap)
cset = ax.contourf(x,y,z, zdir='x', offset=-5, cmap=my_cmap)
cset = ax.contourf(x,y,z, zdir='y', offset=5, cmap=my_cmap)

fig.colorbar(surf, ax=ax, shrink=0.5, aspect=5)

ax.set_xlabel('X-axis')
ax.set_xlim(-5, 5)
ax.set_ylabel('Y-axis')
ax.set_ylim(-5, 5)
ax.set_zlabel('Z-axis')
ax.set_zlim(np.min(z), np.max(z))
ax.set_title('3D Surface Having 2D Contour Plot Projections')

plt.show()
```
