## Working with Images

```python
import matplotlib.pyplot as plt
import matplotlib.image as img

testImage = img.imread('img.jpg')

plt.imshow(testImage)
plt.show()
```

```python
testImage = img.imread('img.jpg')

print(testImage)
```

```python
testImage = img.imread('img.jpg') 
  
# displaying the shape of the image 
print(testImage.shape) 
  
# modifying the shape of the image 
modifiedImage = testImage[:, :, 0] 
  
# displaying the modified image 
plt.imshow(modifiedImage) 
```

```python
testImage = img.imread('img.jpg') 

print(testImage.shape) 

modifiedImage = testImage[50:2000, 1000:3000, 1] 

plt.imshow(modifiedImage) 
```

```python
testImage = img.imread('img.jpg') 

print(testImage.shape) 

modifiedImage = testImage[:,:, :] 

plt.imshow(modifiedImage) 
```

### Aplying Pseudocolor

```python
modifiedImage = testImage[:,:, 0] 

plt.imshow(modifiedImage) 
```

### Another value to color map

```python
modifiedImage = testImage[:,:, 0] 

plt.imshow(modifiedImage, cmap='hot') 
plt.colorbar()
```

### Interpolation Schemes

```python
from PIL import Image 

img = Image.open('img.jpg') 

img.thumbnail((50, 50), Image.ANTIALIAS) 
imgplot = plt.imshow(img) 
```

### Bicubic

```python
img = Image.open('img.jpg') 
  
img.thumbnail((30, 30), Image.ANTIALIAS)  

imgplot = plt.imshow(img, interpolation ='bicubic') 
```

### Sinc

```python
img = Image.open('img.jpg') 
  
img.thumbnail((30, 30), Image.ANTIALIAS)  

imgplot = plt.imshow(img, interpolation ='sinc') 
```
