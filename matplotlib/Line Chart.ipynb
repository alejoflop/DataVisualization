# Line Chart in Patplotlib

Line charts are used to represent the relation between two data X and Y on a different axis.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([1,2,3,4])
y = x*2

plt.plot(x, y)
plt.show()
```

```python
x = np.array([1,2,3,4])
y = x*2

plt.plot(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Chart')
plt.show()
```

### Annotations

```python
x = [1,2,3,4,5]
y = [2,4,6,8,10]

plt.figure(figsize=(8,6))
plt.plot(x, y, marker='o', linestyle='-')

# Add annotations
for i, (xi, yi) in enumerate(zip(x, y)):
    plt.annotate(f'({xi}, {yi})', (xi, yi), 
                 textcoords='offset points', xytext=(0, 10), ha='center')

plt.title('Line Chart with Annotations')
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')

plt.grid(True)
plt.show()
```

### Multiple Line Charts

```python
plt.plot(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Chart')
plt.show()

plt.figure()
x1 = [2,4,6,8]
y1 = [3,5,7,9]
plt.plot(x1, y1, '-.')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title('Multiple Line Chart')
plt.show()
```

### Multiple Plots  on the Same Axis

```python
plt.plot(x, y)

plt.plot(x1, y1, '-.')
plt.xlabel("X-axis data")
plt.ylabel("Y-axis data")
plt.title('Multiple Plots')
plt.show()
```

### Fill the Area Between Two Lines

```python
x = np.array([1, 2, 3, 4])
y = x*2
 
plt.plot(x, y)
 
x1 = [2, 4, 6, 8]
y1 = [3, 5, 7, 9]

plt.plot(x, y1, '-.')
plt.xlabel("X-axis data")
plt.ylabel("Y-axis data")
plt.title('Fill Area')

plt.fill_between(x, y, y1, color='green', alpha=0.5)
plt.show()
```

### Line Plot Styles

```python
import random

students = ["Jane","Joe","Beck","Tom",
            "Sam","Eva","Samuel","Jack",
            "Dana","Ester","Carla","Steve",
            "Fallon","Liam","Culhane","Candance",
            "Ana","Mari","Steffi","Adam"]
 
marks=[]
for i in range(0,len(students)):
     marks.append(random.randint(0, 101))
 
plt.xlabel("Students")
plt.ylabel("Marks")
plt.title("CLASS RECORDS")
plt.plot(students,marks,'m--')
```

### Adding Markers

```python
plt.plot(students, marks, color='g',
         linestyle='solid', marker='o', markerfacecolor='r', markersize=12)
```

### Adding Grid Lines

```python
students = ["Jane", "Joe", "Beck", "Tom",
            "Sam", "Eva", "Samuel", "Jack",
            "Dana", "Ester", "Carla", "Steve",
            "Fallon", "Liam", "Culhane", "Candance",
            "Ana", "Mari", "Steffi", "Adam"]
 
marks = []
for i in range(0, len(students)):
    marks.append(random.randint(0, 101))
 
plt.xlabel("Students")
plt.ylabel("Marks")
plt.title("CLASS RECORDS")
plt.plot(students, marks, 'm--')
 
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.show()
```

## Multiple Lines

### Vertical Line

```python
x = [10,20,30,40,50] 
y = [30,30,30,30,30] 
  
# plot line 
plt.plot(x, y) 
plt.show()
```

### Vertical Line

```python
plt.plot(y,x) 
plt.show()
```

### Horizontal and Vertical Line

```python
plt.plot(x, y, label='line 1')
plt.plot(y, x, label='line 2')
plt.legend()
plt.show()
```

### Plotting Multiple Lines

```python
x = [1,2,3,4,5] 
y = [3,3,3,3,3] 

plt.plot(x, y, label='line 1')
plt.plot(y, x, label='line 2')
plt.plot(x, np.sin(x), label='Curve 1')
plt.plot(x, np.cos(x), label='Curve 2')
plt.legend()
plt.show()
```

### Multiple Lines with Different Line Styles

```python
plt.plot(x, y, label='line 1', linestyle='-')
plt.plot(y, x, label='line 2', linestyle='--')
plt.plot(x, np.sin(x), label='Curve 1', linestyle='-.')
plt.plot(x, np.cos(x), label='Curve 2', linestyle=':')
plt.legend()
plt.show()
```

np.arange(start, end): This function returns equally spaced values from the interval [start, end).
plt.title(): It is used to give a title to the graph. Title is passed as the parameter to this function.
plt.xlabel(): It sets the label name at X-axis. Name of X-axis is passed as argument to this function.
plt.ylabel(): It sets the label name at Y-axis. Name of Y-axis is passed as argument to this function.
plt.plot(): It plots the values of parameters passed to it together.
plt.show(): It shows all the graph to the console.

```python
x = np.arange(1, 11)
y = x*x

plt.title('Line Graph')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.plot(x, y, color='r')
plt.show()
```

```python
x = np.arange(1, 11)
y = np.array([100, 10, 300, 20, 500, 60, 700, 80, 900, 100])

plt.title('Line Graph')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.plot(x, y, color='g')
plt.show()
```
