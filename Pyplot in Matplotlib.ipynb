### Plotting in Matplotlib

```python
import matplotlib.pyplot as plt
 
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
plt.axis([0, 6, 0, 20])
plt.show()
```

### Linear Plot

```python
year = [1972,1982,1992,2002,2012]
e_india = [100.6,158.61,305.54,394.96,724.79]
e_bangladesh = [10.5, 25.21, 58.65, 119.27, 274.87]

plt.plot(year, e_india, color = 'orange', label = 'India')
plt.plot(year, e_bangladesh, color = 'g', label = 'Bangladesh')

plt.xlabel('Years')
plt.ylabel('Power consumption in kWh')

plt.title('Electricity consumption per capita od India and Bangladesh')

plt.legend()
plt.show()
```

### Linear Plot with Line Formatting

Same plot but with a variation in the lines plotted

```python
year = [1972,1982,1992,2002,2012]
e_india = [100.6,158.61,305.54,394.96,724.79]
e_bangladesh = [10.5, 25.21, 58.65, 119.27, 274.87]

plt.plot(year, e_india, color = 'orange', marker = 'o', label = 'India')
plt.plot(year, e_bangladesh, color = 'g', linestyle = 'dashed', label = 'Bangladesh')

plt.xlabel('Years')
plt.ylabel('Power consumption in kWh')

plt.title('Electricity consumption per capita od India and Bangladesh')

plt.legend()
plt.show()
```

### Scatter Plot with Color Mapping

```python
import numpy as np

np.random.seed(42)
x = np.random.rand(50)
y = np.random.rand(50)
colors = np.random.rand(50)

# Create a scatter plor with color mapping
plt.scatter(x, y, c=colors, cmap='viridis', s=100, alpha=0.8)

# Add color bar for reference
cbar = plt.colorbar()
cbar.set_label('Color Value')

# Set plot title and labels
plt.title('Scatter Plor With Color Mapping')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

plt.show()
```
